```python
import seaborn as sns; sns.set()

#Import du dataset
iris = sns.load_dataset("iris")

#On v√©rifie la shape du dataset
iris.shape
```

```python
sns.pairplot(iris, hue='species', corner=True, height=2.3)
```

## Isoler la variable cible des features

```python
Y = iris['species'].astype('category').cat.codes
X = iris.drop('species', axis='columns')
```

```python
X.shape, Y.shape
```

## S√©parer les donn√©es entre TRAINING set & TEST set

```python
from sklearn.model_selection import train_test_split

splits = train_test_split(
    X,
    Y,
    train_size=0.3, #% de valeurs √† mettre dans la partie TEST
    stratify=Y,
    random_state=2)
#On r√©cup√®re un tableau contenant les √©quivalents de X_Train, Y_Train, X_Test, Y_Test

for each  in splits:
    print(each.shape)

X_train, X_test, Y_train, Y_test = splits
```
