```python
joueurs = df_players.dropna()
```

```python
corrJoueurs = joueurs.corr()
corrJoueurs
```

```python
plt.subplots(figsize=(25,20))
sns.heatmap(corrJoueurs, xticklabels=corrJoueurs.columns,yticklabels=corrJoueurs.columns,cmap="hot")
```

```python
sns.catplot(x="From inside the area", y="Shots on target", data=df_players, kind="bar")
```

```python
sns.catplot(x="From inside the area", y="Shots on target", data=df_players.query("`Shots on target` < 41"), kind="bar")
```

```python
sns.boxplot(x="Position", y=df_players, data=df_players)
```

```python
sns.boxplot(data=df_players, x="Position", y="Shots")
```

```python
sns.barplot(data=df_players, x="Team", y=df_players.groupby(['Team'])["Shots"].sum())
```
