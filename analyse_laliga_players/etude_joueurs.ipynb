```python
joueurs = df_players.dropna()
```

```python
corrJoueurs = joueurs.corr()
corrJoueurs
```

```python
plt.subplots(figsize=(25,20))
sns.heatmap(corrJoueurs, xticklabels=corrJoueurs.columns,yticklabels=corrJoueurs.columns,cmap="hot")
```

```python
sns.catplot(x="From inside the area", y="Shots on target", data=df_players, kind="bar")
```

```python
sns.catplot(x="From inside the area", y="Shots on target", data=df_players.query("`Shots on target` < 40"), kind="bar")
```

```python
sns.boxplot(data=df_players.query("(Position != 'Goalkeeper') & ((Position == 'Defender' & `Duels lost` < 115) | (Position != 'Defender')) "), x='Position', y="Duels lost")
```

```python
test = df_players.groupby(['Team'])["Shots"].sum()
df_players.groupby(['Team'])['Team']
```

```python

plt.subplots(figsize=(15,6))
sns.catplot(data=df_players.query("Position == 'Forward' &`Goals scored` > 10"), x="Shots on target", y="Goals scored", hue="Team")
plt.xticks(rotation=90)
plt.title("Buts marqués par tir cadré")
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(data=df_players.query("Position == 'Forward' &`Goals scored` > 5"), x="Team", y="Shots on target")
plt.xticks(rotation=90)
plt.title("Tirs cadrés par équipe")
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.catplot(data=df_players.query("Position == 'Forward' &`Goals scored` > 10"), x="Team", y="Goals scored", hue="Name")
plt.xticks(rotation=90)
plt.title("Joueurs ayant marqué plus de 10 buts")
plt.show()
```
