### 1) Importation de Pandas, Seaborn, Matplotlib, Numpy

```python
import pandas as pd, seaborn as sns, matplotlib as mp, numpy as npy
```

 Import de house pricing

```python
ds = pd.read_csv("csv_files/house_pricing.csv")
```

### 2) Liste des variables

```python
ds.info()
```

### 3) Nuage de points relatif entre LotArea et SalePrice

```python
sns.relplot(data=ds, x="LotArea", y="SalePrice")
```

### 4) Nuage de point affin

```python
sns.relplot(data=ds.query('LotArea < 20000 & SalePrice < 500000'), x="LotArea", y="SalePrice")
```

### 5) Nuage de point relatif entre LotFrontage et LotArea

```python
sns.relplot(data=ds, x="LotFrontage", y="LotArea")
```

### 6) Nuage de points affin

```python
sns.relplot(data=ds.query("LotFrontage < 200 & LotArea < 100000"), x="LotFrontage", y="LotArea")
```
